## Importing the libraries

```
import os
from datetime import datetime 
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.base import BaseEstimator, TransformerMixin
```

## Importing the dataset

```
scores = pd.read_csv('/content/drive/MyDrive/data/scores.csv')
scores = scores[scores.number.str.startswith("condition")]
scores = scores.fillna(0)
scores['edu'] = scores['edu'].apply(lambda x: 0 if x == ' ' else x)
#find average age
agl = []
sum = 0
for a in scores["age"]:
  agl = str(a).split("-")
  length = len(agl)
  for i in range(length):
    sum = sum + float(agl[i])
  av = round (sum / length)
  sum = 0
  scores.age[scores.age == a] = av

scores.head()
```

```
#find average education years
edul = []
sum = 0
for e in scores["edu"]:
  edul = str(e).split("-")
  length = len(edul)
  for i in range(length):
    sum = sum + int(edul[i])
  ave = round (sum / length)
  sum = 0
  scores.edu[scores.edu == e] = ave

scores.head()
```

```
scores = scores[["gender", "age", "afftype", "melanch", "inpatient", "edu", "marriage", "work", "madrs1", "madrs2"]]
scores.head()
```

```
X_1 = scores.loc[:, scores.columns != 'madrs1'].values
# df.loc[:, df.columns != 'b']
y_1 = scores.iloc[:, 8].values
X_2 = scores.loc[:, scores.columns != 'madrs2'].values
y_2 = scores.iloc[:, -1].values
```

## Splitting the dataset into the Training set and Test set

```
from sklearn.model_selection import train_test_split
X_train_1, X_test_1, y_train_1, y_test_1 = train_test_split(X_1, y_1, test_size = 0.08, random_state = 0)
X_train_2, X_test_2, y_train_2, y_test_2 = train_test_split(X_2, y_2, test_size = 0.08, random_state = 0)
```

## Training the SVM model on the Training set

```
from sklearn.svm import SVC
classifier_1 = SVC(kernel = 'linear', random_state = 0)
classifier_1.fit(X_train_1, y_train_1)
classifier_2 = SVC(kernel = 'linear', random_state = 0)
classifier_2.fit(X_train_2, y_train_2)
```

## Making the Confusion Matrix

```
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred_1 = classifier_1.predict(X_test_1)
y_pred_2 = classifier_2.predict(X_test_2)
cm_1 = confusion_matrix(y_test_1, y_pred_1)
cm_2 = confusion_matrix(y_test_2, y_pred_2)
print(cm_1)
accuracy_score(y_test_1, y_pred_1)
```

#### The SVM model achieves an accuracy of 50% which can be further improved upon by using a larger dataset or applying k-fold cross validation
